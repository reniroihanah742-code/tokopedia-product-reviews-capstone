```python
# =============================
#       Import Library
# =============================
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from wordcloud import WordCloud
from langchain_community.llms import Replicate

# =============================
#        Load Dataset
# =============================
df = pd.read_csv("tokopedia-product-reviews-2019.csv")
print(df.shape)
df.head()
```

```python
# =============================
#     Cek Struktur Data
# =============================
print("Ukuran dataset:", df.shape)
print("\nInfo dataset:")
print(df.info())
print("\nContoh data:")
display(df.head())
```

```python
# Panjang review
df["review_length"] = df["text"].astype(str).apply(len)
sns.histplot(df["review_length"], bins=30)
plt.title("Distribusi Panjang Review")
plt.show()
```

```python
# =============================
#  Data Cleaning
# =============================
print("\nMissing values:")
print(df.isnull().sum())

print("\nJumlah duplikat:", df.duplicated().sum())
```

```python
# =============================
# Data Cleaning
# =============================
def label_sentiment(rating):
    if rating >= 4:
        return "positive"
    elif rating == 3:
        return "neutral"
    else:
        return "negative"

df['sentiment'] = df['rating'].apply(label_sentiment)
```

```python
# =============================
# Distribusi Rating
# =============================

sns.countplot(x="rating", data=df, palette="pastel")
plt.title("Distribusi Rating Tokopedia")
plt.show()
```

```python
# =============================
# Wordcloud Review Positif vs Negatif
# =============================

positive_reviews = " ".join(df[df['sentiment']=="positive"]['text'].astype(str))
negative_reviews = " ".join(df[df['sentiment']=="negative"]['text'].astype(str))

plt.figure(figsize=(12,6))
wc_pos = WordCloud(width=800, height=400, background_color="white").generate(positive_reviews)
wc_neg = WordCloud(width=800, height=400, background_color="black").generate(negative_reviews)

plt.subplot(1,2,1)
plt.imshow(wc_pos, interpolation="bilinear")
plt.title("Kata-Kata Umum Review Positif")
plt.axis("off")

plt.subplot(1,2,2)
plt.imshow(wc_neg, interpolation="bilinear")
plt.title("Kata-Kata Umum Review Negatif")
plt.axis("off")
plt.show()
```

```python
from textblob import TextBlob

# Set token Replicate dari Colab Secrets (nama: api_token)
try:
    from google.colab import userdata
    tok = userdata.get("api_token")
    if tok: os.environ["REPLICATE_API_TOKEN"] = tok
except Exception:
    pass

# --- Load data ---
PATH = "tokopedia-product-reviews-2019.csv"   # sesuaikan jika beda
df = pd.read_csv(PATH, low_memory=False)

# --- Otomatis deteksi nama kolom teks & rating ---
def pick(colnames):
    for c in colnames:
        if c in df.columns: return c
    return None

COL_TEXT   = pick(['review','content','text','review_text','comment','ulasan'])
COL_RATING = pick(['rating','stars','rate','score'])

assert COL_TEXT,  f"Tidak menemukan kolom teks ulasan. Kolom tersedia: {list(df.columns)}"
assert COL_RATING,f"Tidak menemukan kolom rating. Kolom tersedia: {list(df.columns)}"

df[COL_TEXT] = df[COL_TEXT].astype(str)
df[COL_RATING] = pd.to_numeric(df[COL_RATING], errors='coerce')
df = df.dropna(subset=[COL_RATING]).reset_index(drop=True)

# --- Label sentimen sederhana dari rating ---
def label_from_rating(r):
    if r >= 4: return "Positif"
    if r <= 2: return "Negatif"
    return "Netral"

df["sentiment_label"] = df[COL_RATING].apply(label_from_rating)

# --- Grafik distribusi sentimen ---
sns.countplot(data=df, x="sentiment_label")
plt.title("Distribusi Sentimen Review")
plt.xlabel("Sentimen"); plt.ylabel("Jumlah")
plt.show()

# --- Siapkan ringkasan numerik untuk LLM ---
n_reviews  = len(df)
avg_rating = df[COL_RATING].mean()
dist_sent  = df["sentiment_label"].value_counts().to_dict()

pos_examples = df.loc[df["sentiment_label"]=="Positif", COL_TEXT].dropna().astype(str).head(3).tolist()
neg_examples = df.loc[df["sentiment_label"]=="Negatif", COL_TEXT].dropna().astype(str).head(3).tolist()

# Opsi tambahan: kalau ingin sentimen berbasis TextBlob (bukan rating)
def blob_polarity(t): return TextBlob(t).sentiment.polarity
# df["sentiment_blob"] = df[COL_TEXT].astype(str).apply(blob_polarity)

# --- Ringkasan AI (Granite via Replicate) ---
from langchain_community.llms import Replicate
llm = Replicate(model="ibm-granite/granite-3.3-8b-instruct")

prompt = f"""
Anda adalah analis data e-commerce. Gunakan ringkasan berikut untuk membuat laporan dalam bahasa Indonesia.

Ringkasan data:
- Jumlah review: {n_reviews}
- Rata-rata rating: {avg_rating:.2f}
- Distribusi sentimen (berdasarkan rating): {dist_sent}
- Contoh review positif: {pos_examples}
- Contoh review negatif: {neg_examples}

Tulis output terstruktur dengan tiga bagian:
1) Hasil analisis — angka kunci & temuan singkat.
2) Wawasan & Temuan — pola utama dari distribusi & contoh.
3) Rekomendasi — langkah konkret untuk seller & platform (bullet points).
"""
report = llm.invoke(prompt)
print(report)
```

```python
from langchain_community.llms import Replicate
from langchain_experimental.agents.agent_toolkits.pandas.base import create_pandas_dataframe_agent

llm = Replicate(model="ibm-granite/granite-3.3-8b-instruct")

agent = create_pandas_dataframe_agent(
    llm=llm,
    df=df,
    verbose=True,
    allow_dangerous_code=True,                  # agent butuh Python REPL
    agent_executor_kwargs={"handle_parsing_errors": True}
)

q1 = (f"Gunakan kolom '{COL_TEXT}' dan '{COL_RATING}'. "
      "Hitung persentase tiap label sentimen (Positif, Netral, Negatif) dan rata-rata rating keseluruhan. "
      "Kembalikan JAWABAN AKHIR dalam bahasa Indonesia sebagai bullet points. Jangan tampilkan Thought/Action.")
out1 = agent.invoke({"input": q1})
print(out1.get("output", out1))

q2 = ("Berdasarkan hasil di atas, berikan 5 rekomendasi bisnis yang praktis untuk seller Tokopedia "
      "(pengemasan, lead time pengiriman, kualitas produk, respons CS, dan manajemen review). "
      "Jawab dalam bullet points.")
out2 = agent.invoke({"input": q2})
print(out2.get("output", out2))
```
